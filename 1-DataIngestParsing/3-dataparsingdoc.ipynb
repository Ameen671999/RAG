### Word Document Processing

```python
from langchain_community.document_loaders import Docx2txtLoader, UnstructuredWordDocumentLoader
```

```python
# Using Docx2txtLoader
print("Using Docx2txtLoader:")

try:
    docx_loader = Docx2txtLoader('data/word_files/Dev II - AI Engineering.docx')
    docx_documents = docx_loader.load()
    print(f'Loaded {len(docx_documents)} documents from DOCX file using Docx2txtLoader.')
    print(f'Doc Content: {docx_documents[0].page_content[:500]}...')  # Print first 500 characters
    print(f'Doc Metadata: {docx_documents[0].metadata}')
except Exception as e:
    print(f"Error loading DOCX with Docx2txtLoader: {e}")
```

```python
## UnstructuredWordDocumentLoader
print("\nUsing UnstructuredWordDocumentLoader:")

try:
    unstructured_loader = UnstructuredWordDocumentLoader('data/word_files/Dev II - AI Engineering.docx', mde="elements")
    unstructured_documents = unstructured_loader.load()
    print(f'Loaded {len(unstructured_documents)} documents from DOCX file using UnstructuredWordDocumentLoader.')
    for i, doc in enumerate(unstructured_documents[:3]):
        print(f'\nDocument {i+1} Content: {doc.page_content[:500]}...')  # Print first 500 characters
        print(f'Document {i+1} Metadata: {doc.metadata.get("category", "unknown"   )}')
        print(f'content: {doc.page_content[:100]}...')  # Print first 100 characters
except Exception as e:
    print(f"Error loading DOCX with UnstructuredWordDocumentLoader: {e}")
```
